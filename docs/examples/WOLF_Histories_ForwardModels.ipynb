# WOLF Histories Forward Models

```python
import pyFTracks
from pyFTracks import Grain
from pyFTracks import Sample
import pyFTracks.thermal_history as TH
from pyFTracks import Viewer
```

```python
sample1 = Sample(Dpars=[1.65])
```

## Ketcham 1999 Annealing model

```python
wolf1 = pyFTracks.Ketcham1999(TH.WOLF1)
wolf2 = pyFTracks.Ketcham1999(TH.WOLF2)
wolf3 = pyFTracks.Ketcham1999(TH.WOLF3)
wolf4 = pyFTracks.Ketcham1999(TH.WOLF4)
wolf5 = pyFTracks.Ketcham1999(TH.WOLF5)
```

```python
Viewer(wolf1, sample1)
```

```python
Viewer(wolf2, sample1)
```

```python
Viewer(wolf3, sample1)
```

```python
Viewer(wolf4, sample1)
```

```python
Viewer(wolf5, sample1)
```
